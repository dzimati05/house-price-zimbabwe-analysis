Data Cleaning 


```python
df.duplicated().sum()
```

Categorial Encoding 

```python
binary_cols = [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea'
]
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
df['furnishingstatus'] = df['furnishingstatus'].map({'furnished': 2, 'semi-furnished': 1, 'unfurnished': 0})
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Graph description 

Are prices evenly distrubuted ??
No because Most of the houses are grouped around the lower to middle price range roughly between : 3m and 6m 


ls the data Skewed ??

Yes: The price distribution is right-skewed. Most houses have lower to moderate prices, while a small number of expensive houses extend the distribution toward the right.


Final EDA interpretation

The distribution of house prices is not evenly spread. Most properties are concentrated between 3 million and 6 million. The graph is right-skewed, meaning there are fewer houses at very high price ranges. A small number of properties above 10 million may be considered expensive outliers. This suggests that the dataset contains mostly mid-range houses, with a few luxury/high-value properties.
